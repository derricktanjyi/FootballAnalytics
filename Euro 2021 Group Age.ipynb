```python
#importing of packages
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import fontawesome as fa

#package to create waffle chart
from pywaffle import Waffle

#package to import highlight text
from highlight_text import fig_text, HighlightText
```

```python
df = pd.read_csv('Euro 2021 Group Age.csv')
df.head()
```

```python
#drop No. of Players Column
df = df.drop('No. of Players',axis=1)

#Transpose the dataframe
df = df.T

#Put Groups at the top as Column Names
df.columns = df.iloc[0]

#drop First Row
df = df.drop('Groups')
```

```python
#transform the dataframe into a list
my_list=df.columns.values.tolist()
df=pd.DataFrame(df, columns=my_list)
df
```

```python
#Plotting for Multiple Waffle Charts
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    icons='',
    figsize=(15,15),
    colors=['#BFAE48','#35A7FF','#FF5964'],
    block_arranging_style='snake',
    block_aspect_ratio=1.05,
    plots = {
        '231' : {
            'values' : df.iloc[:,0],
            'title' : {
                'label' : 'Group A \nItaly, Switzerland, Turkey, Wales',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },         
        '232' : {
            'values' : df.iloc[:,1],
            'title' : {
                'label' : 'Group B \nBelgium, Denmark, Finland, Russia',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },
        '233' : {
            'values' : df.iloc[:,2],
            'title' : {
                'label' : 'Group C \nAustria, Holland, North Macedonia, Ukraine',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },
        '234' : {
            'values' : df.iloc[:,3],
            'title' : {
                'label' : 'Group D \nCroatia, Czech Republic, England, Scotland',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },
        '235' : {
            'values' : df.iloc[:,4],
            'title' : {
                'label' : 'Group E \nPoland, Slovakia, Spain, Sweden',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },
        '236' : {
            'values' : df.iloc[:,5],
            'title' : {
                'label' : 'Group F \nFrance, Germany, Portugal, Hungary',
                'color' : 'black',
                'loc' : 'left',
                'fontdict' : {
                    'fontsize' : 14
                }
            }
        },
        
    }

)

fig.patch.set_facecolor('white')
fig_text(s='Age Distribution of Players in each Euro 2020 Group',
        x=0.21, y=0.91,
        fontsize=23,
        color='#313638'
)
fig_text(s='<Under 23s> | <24-29> | <Over 30s>',
        x=0.32, y=0.88,
        fontsize=20,
        color='#313638',
        highlight_textprops=[{"color":'#BFAE48'},
                              {'color':'#35A7FF'},
                              {'color':'#FF5964'}]
)
```
