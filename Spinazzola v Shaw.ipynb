```python
import matplotlib.pyplot as plt
from highlight_text import fig_text

from mplsoccer import PyPizza, FontManager
```

```python
font_normal = FontManager(("https://github.com/google/fonts/blob/main/apache/arimo/static/"
                           "Arimo-Regular.ttf?raw=true"))
font_italic = FontManager(("https://github.com/google/fonts/blob/main/apache/arimo/static/"
                           "Arimo-Italic.ttf?raw=true"))
font_bold = FontManager(("https://github.com/google/fonts/blob/main/apache/arimo/static/"
                         "Arimo-Bold.ttf?raw=true"))
```

```python
params = ["Expected Assists", "Shot Creation Actions", 
          "Goal Creation Actions", "Pass Completion %",
          "Progressive Passes Distance", "Final 3rd Passes", 
          "Pressures Attempted","Successful Pressure %"]
values = [0.7, 4.05, 0.48, 86.6, 207.1, 4.29, 9.76, 26.8]
values_2 = [1.1, 2.69, 1.08, 84.5, 301.4, 1.35, 12.4, 17.4]
```

```python
min_range = [0, 0.5, 0, 70, 150, 0.5, 3.5, 10]
max_range = [2, 6, 2, 90, 400, 6, 15, 30]
```

```python
# instantiate PyPizza class
baker = PyPizza(
    params=params,
    min_range=min_range,        # min range values
    max_range=max_range,        # max range values
    background_color="#222222", straight_line_color="#000000",
    last_circle_color="#000000", last_circle_lw=2.5, other_circle_lw=0,
    other_circle_color="#000000", straight_line_lw=1
)

# plot pizza
fig, ax = baker.make_pizza(
    values,                     # list of values
    compare_values=values_2,    # passing comparison values
    figsize=(8,8),             # adjust figsize according to your need
    color_blank_space="same",   # use same color to fill blank space
    blank_alpha=0.4,            # alpha for blank-space colors
    param_location=110,         # where the parameters will be added
    kwargs_slices=dict(
        facecolor="#028090", edgecolor="#000000",
        zorder=1, linewidth=1
    ),                          # values to be used when plotting slices
    kwargs_compare=dict(
        facecolor="#FF5D73", edgecolor="#222222", zorder=3, linewidth=1,
    ),                          # values to be used when plotting comparison slices
    kwargs_params=dict(
        color="#F2F2F2", fontsize=13, zorder=5,
        fontproperties=font_normal.prop, va="center"
    ),                          # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=13,
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="#028090",
            boxstyle="round,pad=0.2", lw=1
        )
    ),                           # values to be used when adding parameter-values
    kwargs_compare_values=dict(
        color="#000000", fontsize=12,
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="#FF5D73",
            boxstyle="round,pad=0.2", lw=1
        )
    )                            # values to be used when adding comparison-values
)

# add title
fig_text(
    0.485, 0.99, "<Leonardo Spinazzola> vs <Luke Shaw>",
    size=16, fig=fig,
    highlight_textprops=[{"color": '#028090'}, {"color": '#FF5D73'}],
    ha="center", fontproperties=font_bold.prop, color="#F2F2F2"
)

# add subtitle
fig.text(
    0.515, 0.942,
    "Matches Played: 4 | EURO 2020 | Stats are Per 90mins",
    size=15,
    ha="center", fontproperties=font_bold.prop, color="#F2F2F2"
)

# add credits
CREDIT_1 = "Data: Statsbomb via Fbref"
CREDIT_2 = "Viz Inspired by: @Worville, @FootballSlices, @somazerofc, @Soumyaj15209314"
Twitter = "Twitter: @Pirlo_Azzure"

fig.text(
    0.99, 0.005, f"{CREDIT_1}\n{Twitter}\n{CREDIT_2}", size=10,
    fontproperties=font_italic.prop, color="#F2F2F2",
    ha="right"
)

plt.savefig('SpinazzolaShaw')
plt.show()
```
